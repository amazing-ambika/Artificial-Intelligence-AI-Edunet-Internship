```python
import pandas as pd
import numpy as np
sales_data = {"Person": ["Rajesh","Raju", "Pradeep", "Amit", "Rakesh", "Suresh", "Raj", "Raju"], 
            "Company" : [ "CompA","CompA", "CompA", "CompB", "CompB", "CompC", "CompC", "CompA"],
             "Sales"  : [   250  ,  250  ,   150  ,   340  ,   np.nan ,   243  ,   350  ,   250],
             "Age"    : [    23  ,  24   ,   25   ,   250  ,  np.nan,   27   ,    24  ,    24] }
sales_df = pd.DataFrame(sales_data)
sales_df
```

```python
sales_df.duplicated().sum()
```

```python
# fix duplicates
sales_df.drop_duplicates(inplace=True)
```

```python
sales_df
```

```python
# Handing error
sales_df["Age"].replace(250.0, 25.0, inplace=True)
```

```python
sales_df
```

```python
sales_df.isnull().sum()
```

```python
# Handing missing values
sales_df.dropna()
```

```python
sales_df
```

```python
x=int(sales_df["Age"].mean())
x
```

```python
sales_df["Age"].fillna(x, inplace=True)
```

```python
sales_df
```

```python
x=int(sales_df["Sales"].mean())
```

```python
sales_df["Sales"].fillna(x, inplace=True)
```

```python
sales_df["Sales"].fillna(0, inplace=True)
```

```python
sales_df
```

```python
sales_df.to_csv("Sale Data.csv")
```
